```python
# Largest component (return size of the largest component)

def largest(graph):
    seen = set()
    max = 0
    for node in graph.keys():
        if node not in seen:
            count = 0
            q = deque()
            q.append(node)
            while q:
                cur = q.popleft()
                if cur not in seen:
                    seen.add(cur)
                    count += 1
                for item in graph[cur]:
                    if item not in seen:
                        q.append(item)
            if count > max:
                max = count
    return max


def largest(graph):
    def dfs(graph, cur):
        global seen
        if cur in seen:
            return 0
        
        count = 1
        seen.add(cur)
        for item in graph[cur]:
            count += dfs(graph, item)
        return count
    max = 0
    global seen
    for node in graph.keys():
        if node not in seen:
            count = dfs(graph, node)
            if count > max:
                max = count
    return max
```

```python
# # Finding the shortest path from a source to a destination node
def sPath(graph, src, dst):
    seen = set()
    prev = {}
    q = deque(src)
    while q:
        cur = q.popleft()
        seen.add(cur)
        if cur == dst:
            return prev
        for node in graph[cur]:
            if node not in seen:
                q.append(node)
                seen.add(node)
                prev[node] = cur
    return {}

def printPath(prev, src, dst):
    if not prev:
        return "No path to given destination from the given source"
    path = []
    cur = dst
    while cur != src:
        path.append(cur)
        cur = prev[cur]
    solution = f"{src} -> "
    while len(path) > 1:
        solution += f"{path.pop()} -> "
    solution += path.pop()
    return solution
```

```python
# Find the shortest time to get out the dungeon. (S-E). The time taken should be returned
def checkLimits(grid, r, c):
    if r>= 0 and r< len(grid) and c >= 0 and c < len(grid[0]):
        return True
    else:
        return False
    
def dungeonSolver(dungeon, start):    
    seen = set()
    prev = {}
    q = deque([[start, 0]])
    seen.add(start)
    while q:
        [r, c], count = q.popleft()
        if dungeon[r][c] == 'E':
            return f"The shortest duration to reach the end of the dungeon is: {count} minutes.", prev
        for dr, dc in [(0, 1), (0, -1), (-1, 0), (1, 0)]:
            if checkLimits(dungeon, r+dr, c+dc) and (r+dr, c+dc) not in seen and dungeon[r+dr][c+dc] not in '#':
                seen.add((r+dr, c+dc))
                q.append([(r+dr, c+dc), count+1])
                prev[(r+dr, c+dc)] = (r, c)
    return 'The dungeon cannot be beaten.', {}

def pathMaker(prev, start, end):
    path = ""
    stack = []
    cur = end
    while cur in prev:
        stack.append(cur)
        cur = prev[cur]
    stack.append(start)
    while len(stack) > 1:
        path += f"{stack.pop()} -> "
    path += f"{end}"
    return path
```

```python
# SSSP on DAG (Single source shortest path)
import math

# Shortest path
def relaxation(graph, cur, order, distances, index):
    for node, cost in graph[cur]:
        if cost + distances[index] < distances[order.index(node)]:
            distances[order.index(node)] = cost + distances[index]
    return distances

# Longest path.
def relaxation(graph, cur, order, distances, index):
    for node, cost in graph[cur]:
        if (cost * -1) + distances[index] < distances[order.index(node)]:
            distances[order.index(node)] = (cost * -1) + distances[index]
    return distances

def shortestPath(graph):
    n = len(graph.keys())
    distances = [math.inf] * n
    order = topological(graph)
    distances[0] = 0
    
    for index, node in enumerate(order):
        distances = relaxation(graph, node, order, distances, index)
    return order, [item * -1 for item in distances]

# To find the longest path, we just multiply all costs by -1, find the min distance and multiply that by -1
```

```python
# Dijkstra Algorithm (Lazy approach)

# Essentially, start at a given node, proceed to mark visited node as seen, optimize the value for every node and try finding the best path from source to destination
from math import inf

def dijkstra(graph, src, dst):
    seen = set()
    pq = deque()
    distances = [inf] * len(graph.keys())
    prev = {}
    distances[src] = 0 
    pq.append((src, distances[src]))
    while pq:
        node, till = pq.popleft()
        if node == dst:
            path = pathMaker(prev, src, dst)
            return distances[dst], path
        seen.add(node)
        for item, cost in graph[node]:
            if till + cost < distances[item]:
                prev[item] = node
                distances[item] = till + cost
            pq.append((item, till + cost))
        pq = deque(sorted(pq, key=lambda x: x[1]))
    return "Node not found", {}
```

```python
def bellman_ford(graph):
    n = len(graph.keys())
    distances = [inf] * n
    distances[0] = 0
    for _ in range(0, n-1):
        for node in graph.keys():
            for neighbour, cost in graph[node]:
                if distances[node] + cost < distances[neighbour]:
                    distances[neighbour] = distances[node] + cost
    
    for _ in range(0, n-1):
        for node in graph.keys():
            for neighbour, cost in graph[node]:
                if distances[node] + cost < distances[neighbour]:
                    distances[neighbour] = -inf
    return distances
```

```python
# Floyd Warshall Algorithm
# Essentially, try finding the APSP (All pairs shortest path from each node)
# to all other nodes.
def nextMat(graph, matrix, n):
    for i in range(n):
        for j in range(n):
            if graph[i][j] != inf:
                matrix[i][j] = j
    return matrix

def negativeLoopFinder(graph, n, matrix):
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if i != k and j != k and i != j:
                    if graph[i][j] > graph[i][k] + graph[k][j]:
                        graph[i][j] = -inf
                        matrix[i][j] = -1

    return graph, matrix
                   
def floyd_warshall(graph):
    n = len(graph)
    next = [[None] * n for _ in range(n)]
    next = nextMat(graph, next, n) 
    for k in range(n): # For each of the nodes
        for i in range(n):
            for j in range(n): # iterating through the graph
                if i != k and j != k and i != j:
                    if graph[i][k] + graph[k][j] < graph[i][j]:
                        graph[i][j] = graph[i][k] + graph[k][j]
                        next[i][j] = next[i][k]
    graph, next = negativeLoopFinder(graph, n, next)
    return graph, next

def findPath(graph, next, start, end):
    path = []
    if graph[start][end] == inf:
        return path
    
    cur = start
    while cur != end:
        if cur == -1:
            return None
        path.append(cur)
        cur = next[cur][end]

    if next[cur][end] == -1:
        return None
    path.append(end)
    return path



# For finding any negative cycles, all we have to do is re-run the algorithm
# n times. If any of the values are updated to new ones, there exists a neg-
# ative cycle. (Similar to Bellman Ford)
```
